# 02 Deriving Variables (Lesson)

```python
# define a new column
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# check the output
df.head()
```

```python
# count the outputs in our new column
df['price_range'].value_counts()
```

```python
# check the maximum price
df['prices'].max()
```

```python
# do the same thing using the loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# run the same function on the entire data frame
df_merged.loc[df_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_merged.loc[(df_merged['prices'] <= 15) & (df_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

#### Question 4 - Create a 'busiest_period_of_day' column

```python
df_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create the 'busiest_period_of_day' column with the loc function
# Most orders-- hours 9-17
# Average orders-- hours 7-8, 18-23
# Fewest orders-- hours 0-6

df_merged.loc[(df_merged['order_hour_of_day'] <= 17) & (df_merged['order_hour_of_day'] >= 9), 'busiest_period_of_day'] = 'Most orders'
```

```python
df_merged.loc[(df_merged['order_hour_of_day'] <= 23) & (df_merged['order_hour_of_day'] >= 18), 'busiest_period_of_day'] = 'Average orders'
```

```python
df_merged.loc[(df_merged['order_hour_of_day'] >= 7) & (df_merged['order_hour_of_day'] <= 8), 'busiest_period_of_day'] = 'Average orders'
```

```python
df_merged.loc[df_merged['order_hour_of_day'] < 7, 'busiest_period_of_day'] = 'Fewest orders'
```
