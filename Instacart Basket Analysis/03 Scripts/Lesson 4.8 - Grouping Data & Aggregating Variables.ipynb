# 02 Grouping and Aggregating Data

```python
# try the groupby() function (split the data into groups)
df.groupby('product_name')
```

```python
# aggregate the data
df.groupby('department_id').agg({'number_of_orders_by_user': ['mean']})
```

```python
# same results, different method
df.groupby('department_id')['number_of_orders_by_user'].mean()
```

```python
# We can also use this code, but it is better to use brackets for indexing!
df.groupby('department_id').number_of_orders_by_user.mean()
```

```python
# multiple aggregations

df.groupby('department_id').agg({'number_of_orders_by_user': ['mean', 'min', 'max']})
```

```python
# criteria for our flag:

# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# ...over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# ...less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
# create a new max_order column
df_merged['max_order'] = df_merged.groupby(['user_id'])['number_of_orders_by_user'].transform(np.max)
```

```python
# check the output
df_merged.head(15)
```

```python
# change the settings so Jupyter displays all the rows you want

pd.options.display.max_rows = None
```

```python
# check the output

df_merged.head(100)
```

```python
# create a flag column

df_merged.loc[df_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_merged.loc[(df_merged['max_order'] <= 40) & (df_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_merged.loc[df_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
